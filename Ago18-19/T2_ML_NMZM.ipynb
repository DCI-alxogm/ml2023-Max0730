```python
data = pd.read_csv('/content/drive/MyDrive/Data/anime.csv')
```

```python
plt.scatter(data['rating'], data['members'], color = 'purple', s = 10, alpha=0.4)
plt.title('Rating de anime vs Cantidad de votos')
plt.xlabel('Rating')
plt.ylabel('Votos');
```

```python
def min_cuadrados(x, y):
    N = len(x)
    m = (np.sum(x*y) - np.sum(x)*np.sum(y)/N)/(np.sum(x*x) - (np.sum(x)**2)/N)
    b = (np.sum(y) - m*np.sum(x))/N

    return (m, b)

x = data['rating']
y = data['members']

slope, b0 = min_cuadrados(x, y)

x1 = np.linspace(np.min(x), np.max(x), 250)
```

```python
plt.scatter(data['rating'], data['members'], color = 'purple', s = 10, alpha = 0.4)
plt.plot(x1, slope*x1+ b0, color = 'blue', ls = '--')
plt.title('Votos vs Rating')
plt.xlabel('Rating')
plt.ylabel('Votos');
```
