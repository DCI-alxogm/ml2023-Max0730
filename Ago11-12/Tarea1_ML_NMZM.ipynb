```python
X = rand.rand(23, 2)
plt.scatter(X[:, 0], X[:, 1], s=30, c='brown', alpha=0.6);  #plotea los puntos

differences = X[:, np.newaxis, :] - X[np.newaxis, :, :] #para cada par de puntos, establece diferencias en sus coordenadas
sq_differences = differences ** 2
dist_sq = sq_differences.sum(-1)  #Suma las diferencias entre coordenadas para tener una distancia al cuadrado
K = 2
nearest_partition = np.argpartition(dist_sq, K + 1, axis=1)
for i in range(X.shape[0]):
    for j in nearest_partition[i, :K+1]:
        plt.plot(*zip(X[j], X[i]), color='black', alpha=0.3)
        #Traza líneas entre un punto y sus vecinos próximos (se repiten)
```

```python
# Creamos una distribución normal de datos con valores aleatorios
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

#Colocamos puntos en el gráfico principal
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.3, c='blue')

#Creamos un histograma para los datos de cada eje
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='yellow')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='pink')
y_hist.invert_xaxis()
```
